```python
import pandas as pd 
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

data = pd.read_csv("activities.tsv",sep = '\t')

facility_type_counts = data['facility_type'].value_counts()

# Create a bar plot
facility_type_counts.plot(kind='bar', color='skyblue', edgecolor='black')

# Adding labels and title
plt.xlabel('Facility Type')
plt.ylabel('Count')
plt.title('Counts of Facility Types')

# Rotating x-axis labels for better visibility
plt.xticks(rotation=90, ha='right')

# Display the plot
plt.show()
```

```python
park_data = pd.read_csv("activities.tsv", sep='\t')

# Get the counts of each neighborhood and sort them in descending order
neighborhood_counts = park_data['center'].value_counts().sort_values(ascending=False)

# Increase figure size
plt.figure(figsize=(10, 6))

# Adjust font size
plt.rcParams['xtick.labelsize'] = 8  # Adjust the font size as needed

# Create a line plot using the DataFrame directly
plt.plot(neighborhood_counts.index, neighborhood_counts.values, marker='o', linestyle='-', color='skyblue')

# Adding labels and title
plt.xlabel('Park')
plt.ylabel('Number of Facilities')
plt.title('Number of Facilities in Each Park')

# Rotating x-axis labels for better visibility
plt.xticks(rotation=90, ha='right')

# Improve layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python
park_data = pd.read_csv("parks.tsv", sep='\t')

# Get the counts of each neighborhood and sort them in descending order
neighborhood_counts = park_data['neighborhood'].value_counts().sort_values(ascending=False)

# Create a bar plot using the DataFrame directly
plt.bar(neighborhood_counts.index, neighborhood_counts.values, color='skyblue', edgecolor='black')

# Adding labels and title
plt.xlabel('Neighborhood')
plt.ylabel('Count')
plt.title('Parks Count by Neighborhood')

# Rotating x-axis labels for better visibility
plt.xticks(rotation=90, ha='right',ticks=np.arange(0, len(neighborhood_counts), 2))

# Display the plot
plt.show()
```
