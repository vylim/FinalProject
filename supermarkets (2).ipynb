#### Number of Supermarkets in each neighborhood



Another factor taken into consideration was the number of supermarkets in each neighborhood. The best neighborhood was determined by which neighborhood had the most supermarkets. This was considered a factor because when a family decides which neighborhood to live in, supermakrets to buy grocery's are very important because no one wants to travel long distances when going to buy food each weekend. In addition, with multiple supermarkets, it gives families a variety of choices to pick from. 
    
In order to narrow down the data, the neighborhoods with <1 supermarkets were removed. Also, when I was reading the csv file, I realized that many Pittsburgh neighborhoods were labeled as just "pittsburgh", so I made a new csv file replacing all of them with their respective neighborhood names. Another problem I ran into was that this data set also includes suburbs which are not listed as a "neighborhood" in pgh and they were really far from the city, so I just ended up removing them from the data set when I filtered the data to see how many supermarkets in each neighborhood.

```python
import pandas as pd
%matplotlib inline
 ##import original data
super_csv = "https://data.wprdc.org/dataset/cd2b3e27-ca31-43e0-a8c6-2e6c43b4050a/resource/6f18890b-b8f7-4bc8-8086-f5401b6e086c/download/geocodedfoodfacilities-filtered-supermarkets-v2.csv"
super_data = pd.read_csv(super_csv)
super_data.head(5)
```

```python
##filtered data set the only the data I am working with
super_data[['name', 'street_address', 'city']].head(20)
```

```python
#group data to only show the city and number of grocery stores it has
#plotting the data on a bar graph
import matplotlib.pyplot as plt
plt.bar(super_data["city"], super_data["name"])
plt.xlabel("Neighborhood")
plt.ylabel("Number of Supermarkets")
plt.title("Number of Supermarkets in Each Neighborhood")
plt.xticks(rotation=45, ha="right")  
plt.tight_layout()  

plt.show()
```

Graphing the original data, it is very hard to read so I am going to group together the number of supermarkets in each neighborhood so the graph is nicer. In addition, I am going to get rid of supermarkets with <1 so I can find the top number more easily.

```python
import pandas as pd

# Load the new dataset from "new_supermarkets.csv"
new_supermarket_data = pd.read_csv("new_supermarkets.csv")

##exclude cities that are far away and not considered a neighborhood
excluded_cities = ["Wexford", "West Mifflin", "McKeesport", "Verona", "Pleasant", "Gibsonia", "Natrona Heights",
                   "Whitehall", "Ross", "Monroeville"]

# Remove cities from the dataset
filtered_supermarket_data = new_supermarket_data[~new_supermarket_data["city"].isin(excluded_cities)]

#group the data so it counts up how many supermarkets are in each city
result = filtered_supermarket_data.groupby("city")["name"].count()

# Sort the results 
sorted_result = result.sort_values(ascending=False)

##SPECIAL NOTE: all neighborhoods with less than 1 supermarket was removed from the data set to find out top neighborhoods easier
filtered_result = sorted_result[sorted_result > 1]

ranked_cities = filtered_result.reset_index()
top_cities = ranked_cities.head(24)
print(top_cities)

```

```python
#plotting the data on a bar graph
import matplotlib.pyplot as plt
plt.bar(ranked_cities["city"], ranked_cities["name"])
plt.xlabel("Neighborhood")
plt.ylabel("Number of Supermarkets")
plt.title("Number of Supermarkets in Each Neighborhood")
plt.xticks(rotation=45, ha="right")  
plt.tight_layout()  

plt.show()
```

Now, with all the cities with less than one supermarket eliminated, it makes the graph more simplier to read and find out the top neighborhoods.
